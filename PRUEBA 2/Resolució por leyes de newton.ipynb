# Resolución por Leyes de Newton

Como se pudo ver el ejercicio planteado tiene 3 grados de libertad, para resolver por leyes de Newton necesitamos hacer un DCL para las 3 ecuaciones de movimiento que se deben hallar para eso se tiene los siguiente DCL's, las coordenadas que se utilizaron fueron transversal 
$\theta,x_{1},x_{2}$.

## BARRA (m,L)

![title](G3_E1_DCLbarra1.png)

La barra L, se mueve alrededor del punto A, este sistema se puede tomar como inercial. La barra posee un movimiento netamente angular para eso planteemos la ecuación de torques con respecto al punto A para la barra rígida tomando en cuenta el DCL #1.

Con respecto a las elongaciones se tienen las siguientes expresiones para el sistema, deducidas a partir del arco infinitesimal que recorre la barra y la longitud que recorran los cuerpos m y 2m:

$$x'= L\theta $$

$$x'_{1} = x_{1}-\frac{2L}{3}L\theta$$

$$x'_{2} = x_{2}-x_{1}$$

La Inercia de la barra con respecto al punto A es $I_{A}=\frac{1}{3}mL^2$

$$\vec{\sum \tau_{/A}}= I_{/A}\vec{\ddot{\theta}}= \vec{\frac{\partial H_{A}}{\partial t}} + \vec{v_{A/O}} \times \vec{P_{G/O}}$$

Podemos ver que el término $\vec{v_{A/O}}\times \vec{P_{G/O}}$ no existe.

$$Kx_{1}'\frac{2L}{3}-Kx'L = I_{/A}\vec{\ddot{\theta}}$$

Reemplazando las elongaciones:

$$ K\left(x_{1}-\frac{2L}{3}L\theta\right)\frac{2L}{3}-K(L\theta)L=\frac{1}{3}mL^2\theta$$

$$ KL^2\theta\left(-\frac{4}{9}-1\right)+\frac{2L}{3}Kx_{1}=\frac{1}{3}mL^2\theta$$

Finalmente se tiene la primera ecuación de movimiento del sistema que define el mov. de la barra:

$$\frac{1}{3}mL^2\theta+\frac{13}{9}KL^2\theta-\frac{2}{3}KLx_{1}=0$$

## MASA (m)

![title](G3_E1_DCLm1.png)

La masa (m) se traslada en $x_{1}$, entonces para hallar la ecuación de movimiento de este cuerpo, realicemos una sumatoria de fuerzas en este eje tomando en cuenta el DCL #2:

$$\vec{\sum F_{x_{1}}}= m\vec{\ddot{x_{1}}}$$

$$ 2K(x'_{2})-k(x'_{1})=m\ddot{x_{1}}$$

$$ 2K(x_{2}-x_{1})-k(x_{1}-\frac{2L}{3}L\theta)=m\ddot{x_{1}}$$

$$2kx_{2}-2kx_{1}-kx_{1}+k\frac{2}{3}L\theta=m\ddot{x_{1}}$$

Finalmente, obtenemos la ecuación de movimiento que define a la masa m:

$$m\ddot{x_{1}}-\frac{2}{3}KL\theta+3kx_{1}-2kx_{2}=0$$

## MASA (2m)

![title](G3_E1_DCLm2.png)

La masa (2m) se traslado en $x_{2}$, de la misma forma que con m, realicemos una sumatoria de fuerza en el eje $x_{2}$ como en el DCL #3 para obtener su ecuación de movimiento:

$$\vec{\sum F_{x_{2}}}= 2m\vec{\ddot{x_{2}}}$$

$$-2Kx_{2}'+F(t)=2m\ddot{x_{2}}$$

$$-2K(x_{2}-x_{1})+F(t)=2m\ddot{x_{2}}$$

$$-2Kx_{2}+2Kx_{1}+F(t)=2m\ddot{x_{2}}$$

Finalmente, obtenemos la ecuación de movimiento que define a la masa 2m:

$$2m\ddot{x_{2}}-2Kx_{1}+2Kx_{2}=F(t)$$



## Resolución por Lagrange

Verificamos que nuestro sistema de referencia cumpla con los siguientes requisitos:

-Holonómico.
-Completo. 
-Independiente. 

Tercero: Analizamos T y V para cada cuerpo rígido: 

Para la energía cinética del sistema se tiene:

$$T=\frac{1}{2}m \left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2$$ 

    Para la energía potencial del sistema se tiene:
 
$$V=\frac{1}{2}k(L\theta)^2+\frac{1}{2}k\left(x_1-\frac{2}{3}L\theta\right)^2+\frac{1}{2}2k(x_2-x_1)^2$$ 
   
Cuarto: Procedemos a determinar $(1), (2) y (3)$:

$$(1): \frac{d}{dt}\left(\frac{\partial T}{\partial\dot{\theta}}\right)=\frac{d}{dt}\left(\frac{\partial}{\partial\dot{\theta}}\left(\frac{1}{2}m\left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2\right)\right)$$ 

$$\frac{d}{dt}\left(\frac{\partial T}{\partial\dot{\theta}}\right)=\frac{d}{dt}\left(\frac{1}{3}mL^2\dot{\theta}\right)$$ 

$$\frac{d}{dt}\left(\frac{\partial T}{\partial\dot{\theta}}\right)=\frac{1}{3}mL^2\ddot{\theta}$$



$$(1): \frac{d}{dt}\left(\frac{\partial T}{\partial\dot{x_1}}\right)=\frac{d}{dt}\left(\frac{\partial}{\partial\dot{x_1}}\left(\frac{1}{2}m\left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2\right)\right)$$ 


$$\frac{d}{dt}\left(\frac{\partial T}{\partial\dot{x_1}}\right)=\frac{d}{dt}(m\dot{x_1})$$

$$\frac{d}{dt}\left(\frac{\partial T}{\partial\dot{x_1}}\right)=m\ddot{x_1}$$ 


$$(1): \frac{d}{dt}\left(\frac{\partial T}{\partial\dot{x_2}}\right)=\frac{d}{dt}\left(\frac{\partial}{\partial\dot{x_2}}\left(\frac{1}{2}m\left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2\right)\right)$$ 

$$\frac{d}{dt}\left(\frac{\partial T}{\partial\dot{x_1}}\right)=\frac{d}{dt}(2m\dot{x_2})$$ 

$$\frac{d}{dt}\left(\frac{\partial T}{\partial\dot{x_1}}\right)=2m\ddot{x_2}$$ 


$$(2): \frac{\partial T}{\partial\theta}=\frac{\partial}{\partial\theta}\left(\frac{1}{2}m\left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2\right)$$

$$\frac{\partial T}{\partial\theta}=0$$ 


$$(2): \frac{\partial T}{\partial x_1}=\frac{\partial}{\partial x_1}\left(\frac{1}{2}m\left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2\right)$$ 

$$\frac{\partial T}{\partial x_1}=0$$ 


$$(2): \frac{\partial T}{\partial x_2}=\frac{\partial}{\partial x_2}\left(\frac{1}{2}m\left(\frac{L}{2}\dot{\theta}\right)^2+\frac{1}{2}\frac{1}{12}mL^2\dot{\theta}^2+\frac{1}{2}m\dot{x_1}^2+\frac{1}{2}2m\dot{x_2}^2\right)$$ 

$$\frac{\partial T}{\partial x_2}=0$$ 

$$(3): \frac{\partial V}{\partial\theta}=\frac{\partial}{\partial\theta}\left(\frac{1}{2}k(L\theta)^2+\frac{1}{2}k\left(x_1-\frac{2}{3}L\theta\right)^2+\frac{1}{2}2k(x_2-x_1)^2\right)$$

$$\frac{\partial V}{\partial\theta}=\frac{13}{9}kL^2\theta -\frac{2}{3}kx_{1}L$$


$$(3): \frac{\partial V}{\partial x_1}=\frac{\partial}{\partial x_1}\left(\frac{1}{2}k(L\theta)^2+\frac{1}{2}k\left(x_1-\frac{2}{3}L\theta\right)^2+\frac{1}{2}2k(x_2-x_1)^2\right)
\frac{\partial V}{\partial x_1}=-\frac{2}{3}kL^2\theta+3kx_1-2kx_2$$ 


$$(3): \frac{\partial V}{\partial x_2}=\frac{\partial}{\partial x_2}\left(\frac{1}{2}k(L\theta)^2+\frac{1}{2}k\left(x_1-\frac{2}{3}L\theta\right)^2+\frac{1}{2}2k(x_2-x_1)^2\right)$$ 

$$\frac{\partial V}{\partial x_2}=-2kx_1+2kx_2$$

Quinto: Hallamos (4):

$$\sum\vec{F_i}\cdot\vec{dr}=F(t)$$ 

Sexto: Precedemos a realizar la respectiva suma (1)+(2)+(3)=(4) para determinar la ecuacion de movimiento del sistema por el método de Lagrange:

$$\left(\frac{1}{3}mL^2\ddot{\theta}+m\ddot{x_1}+2m\ddot{x_2}\right)+(0)+\left(\frac{13}{9}kL^2\theta -\frac{2}{3}kx_1L-\frac{2}{3}kL^2\theta+3kx_1-2kx_2-2kx_1+2kx_2\right)=F(t)$$ 


$$\left(\frac{1}{3}mL^2\ddot{\theta}+m\ddot{x_1}+2m\ddot{x_2}\right)+\left(\frac{13}{9}kL^2\theta -\frac{2}{3}kx_1L-\frac{2}{3}kL^2\theta+3kx_1-2kx_2-2kx_1+2kx_2\right)=F(t)$$


$$\left(\frac{1}{3}mL^2\ddot{\theta}+m\ddot{x_1}+2m\ddot{x_2}\right)+\left(\frac{7}{9}kL^2\theta+kx_1\left(\frac{3-2L}{3}\right)\right)=F(t)$$

## Explicación de diferencias de métodos en la dificultad o facilidad de resolución

Para la resolución por leyes de Newton se analizaron los cuerpos uno por uno, además de realizar los DCL'S de cada uno de ellos. Una ventaja que se encontró fue que el segundo término que acompaña a la variación de cantidad de movimiento angular se hace cero haciendo por leyes de Newton.

Un aspecto similar en ambos métodos que resulta un poquito tedioso es encontrar las relaciones de la elongación con respecto al arco que recorre la barra. Estos términos se ocuparon en ambos métodos de resolución.

En general, resulta ser un movimiento complejo del sistema ya que las aceleraciones angulares como las traslacionales x1,x2 dependen del arco que recorre la barra al girar y de las elongaciones de los resortes y son independientes. Es por ello que no se optó por realizar un gráfico.

Con respecto a la resolución por Lagrange, se facilita trabajar por energías ya que al considerar la energía potencial gravitacional nula por las consideraciones planteadas, lo más tedioso llega ser encontrar las relaciones de elongación que también se debían usar en el otro método.
